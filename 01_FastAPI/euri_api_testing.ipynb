```python
import requests

def generate_completion():
    url = "https://api.euron.one/api/v1/euri/chat/completions"
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {os.environ.get('EURI_API_KEY', 'your-api-key-here')}"
    }
    payload = {
        "messages": [
            {
                "role": "user",
                "content": "What's going on between United States of America and Iran ?"
            }
        ],
        "model": "gpt-4.1-nano",
        "max_tokens": 1000,
        "temperature": 0.7
    }

    response = requests.post(url, json=payload, headers=headers)
    data = response.json()
    print(data)

generate_completion()
```

```python
import requests

url = "https://api.euron.one/api/v1/euri/images/generations"

payload = {
    "model": "gpt-4.1-nano",
    #"model": "black-forest-labs/FLUX.1-schnell",
    "prompt": "Goku vs Vegeta in a battle scene, highly detailed, dynamic lighting, vibrant colors, cinematic composition",
    "n": 1,
    "size": "1024x1024",
    "quality": "hd",
    "response_format": "url"
}
headers = {
    "Authorization": f"Bearer {os.environ.get('EURI_API_KEY', 'your-api-key-here')}",
    "Content-Type": "application/json"
}

response = requests.post(url, json=payload, headers=headers)

print(response.text)
```
