#### Understanding the Terms:

##### 1. Vector:
- ordered collection of numbers. 
```
Examples -> [10, 20] 
            [5, 8, 12]
            [-0.0324, -0.0084, 0.1234, ...]
```
- Suppose I want to describe a person through the following attributes -> Height, Weight, Age

    I can represent that person as a vector like this, for example : [180, 75, 30]
- In AI, we represent many things as vectors:
    - Images
    - Audio
    - Text
    - Videos

##### 2. Embedding:
- process of converting something (text, image, audio, etc.) into numbers.
- Example:
```
Text -> I love dogs
Embedding model converts it into -> [0.23, -0.44, 0.11, ...]
```
- Why do we need embeddings?
    - Because Computers don't understand "I love dogs", they only understand -> [0.23, -0.44, 0.11, ...]
    - Therefore, 
            ```
                Text -> Embedding Model -> Numbers
            ```

##### 3. Embedding Vector:
- output of an embedding model is called an embedding vector.
- Example: [0.23, -0.44, 0.11, ...]

##### 4. Dimension:
- one position in a vector
- Example: [10, 20]
    - Dimension 1 = 10
    - Dimension 2 = 20

##### 5. Vector Space:
- universe where vectors live.
- Example: 
```
2-D Space for [3,5] means x = 3 and y = 5

Embedding space for [1536 numbers] means -> 
dimension 1
dimension 2
dimension 3
...
dimension 1536
```

##### 6. 1536-Dimensional Vector Space:
- Suppose:
```
embedding = [
  0.12,
  -0.33,
  0.56,
  ...
]
```
contains 1536 numbers. Then len(embedding) returns 1536 which means the vector has 1536 dimensions.

##### 7. Learned Numerical Feature:
- When the embedding model was trained, it read billions of sentences. During training, it learned patterns such as:
    - positivity
    - negativity
    - food-relatedness
    - animal-relatedness
    - sports-relatedness
    - programming-relatedness
and thousands of other hidden patterns.

- Each dimension begins capturing some learned characteristic. For example: [0.45, -0.21, 0.91, ...]
- Imagine:
```
    Dimension 1 → positivity
    Dimension 2 → food-relatedness
    Dimension 3 → animal-relatedness
```
In reality, we don't know exactly what each dimension means. The model learned them automatically. Therefore they are called: Learned Numerical Features because the model learned them during training.

##### 8. Semantic Meaning of the Text:
- Semantic -> means 'Meaning'
- When we talk about the semantic meaning of a sentence, we are asking: "What is this sentence trying to say?" not "What exact words were used?"
- "Semantic Meaning of the Text" means what is the text trying to say.

````python
# linalg.norm() calculates the Euclidean distance between two vectors. The smaller the distance, the more similar the vectors are.
# what does inp-a or inp-b or inp-c mean ? It calculates the difference between the two vectors namely inp and a, inp and b, inp and c respectively.
# The result of inp-a is a new vector that represents the difference between the two vectors. The Euclidean distance is then calculated from this 
# difference vector, which gives a measure of how far apart the two original vectors are in the vector space. A smaller distance indicates that 
# the vectors are more similar, while a larger distance indicates that they are more dissimilar.
dist_inp_a = np.linalg.norm(inp-a)
dist_inp_b = np.linalg.norm(inp-b)
dist_inp_c = np.linalg.norm(inp-c)
````

````python
dist_inp_a , dist_inp_b, dist_inp_c

# According to the below output, we can conclude that the input "What do you do ?" is more similar to the input 
# "Rakesh is a software engineer with 5 years of experience in web development." because the distance between their embeddings (dist_inp_b) is 
# smaller than the distances to the other two inputs (dist_inp_a and dist_inp_c)
````
